# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# get matplot to show charts inline
%matplotlib inline
````

````python
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

**1. Find Dimension of Array**

Find dimension of each array: height, weight, population
```
# Expected Ouput
Dimension(height): 1
Dimension(weight): 1
Dimension(population): 2
```

````python
print(height.ndim)
print(weight.ndim)
print(population.ndim)
````

**2. Find Shape of Array**

Find shape of each array: height, weight, population
```
# Expected Ouput
Shape(height) : (7000,)
Shape(weight) : (7000,)
Shape(population) : (7000, 2)
```



````python
print(np.shape(height))
print(np.shape(weight))
print(np.shape(population))
````

**3. Find Size of Array**

Find shape of each array: height, weight, population
```
# Expected Ouput:
Size(height) : 7000
Size(weight) : 7000
Size(weight) : 14000
```



````python
#total number of rows + total number od collumns
print(np.size(height))
print(np.size(weight))
print(np.size(population))
````

**8. Create a histogram of height and weight**

````python
plt.hist(height)
plt.xlabel('Height')
plt.ylabel('No. of employees')
plt.show()
````

````python
plt.hist(weight, color='r')
plt.xlabel('Weight')
plt.ylabel('No. of employees')
plt.show()
````

````python
fig, ax = plt.subplots(2,1,sharey=True)
ax[0].hist(height)
ax[1].hist(weight, color='r')
ax[0].set_xlabel('Height')
ax[1].set_xlabel('Weight')
plt.show()
````

**9. Create a boxplot of height and weight**

````python
plt.boxplot(height)
plt.show()
````

````python
plt.boxplot(weight)
plt.show()
````

````python
fig, ax = plt.subplots(1,2,sharey=True)
ax[0].boxplot(height)
ax[1].boxplot(weight)
ax[0].set_xlabel('Height')
ax[1].set_xlabel('Weight')
````
