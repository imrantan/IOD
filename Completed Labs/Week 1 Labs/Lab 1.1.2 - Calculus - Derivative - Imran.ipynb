# Calculus - Derivative

```python
import numpy as np
import matplotlib.pyplot as plt

# get matplot to show charts inline
%matplotlib inline
```

```python
# define a function such as f(x) = x**3 - attempt a variety of different functions
def f(x):
    return x**3
```

```python
# test the function you defined above by simply print its value for some numbers
print('f(2) = ',f(2))
print('f(5) = ',f(5))
print('f(-3) = ',f(-3))
```

```python
# plot f(x)
x = np.arange(-10,11,1)
y = f(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```

```python
# define derivative function
def derivative(x):
    # set delta (change in x) to a small value, say 0.001  --- this to approximate a point
    delta = 1.0/1000.0
    # define rise (the change in f(x)) --- checking the diff in y value of the two points which are very close to each other
    rise = (f(x + delta)) - f(x)
    slope = rise / delta
    return slope
```

```python
# test the derivative function for some values of x
print('derivative of f(2) = ', derivative(2))
print('derivative of f(-7) = ', derivative(-7))
```

```python
from sympy import *

# define a function such as f(x) = x**3 - attempt a variety of different functions
def f(x):
    return x**3

#def a function using Sympy to calculate the derivative
def method(a):
    x= Symbol('x')
    y= f(x)
    deriv= Derivative(y, x)
    z = deriv.doit().subs({x:a})
    return z

# test the derivative function for some values of x
print('derivative of f(2) = ', method(2))
print('derivative of f(-7) = ', method(-7))
```

```python
# plot derivative of f(x)
x = np.arange(-10,11,1)
y = derivative(x)

fig, ax1 = plt.subplots(figsize = (8,5))

color = 'tab:cyan'
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)', color=color)
ax1.plot(x, y, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(which='major', axis='both', linestyle='--')
plt.show()
```

```python
# plot both x and derivative of f(x)
x = np.arange(-10,11,1)
y = f(x)

fig, ax1 = plt.subplots(figsize = (15,10))

color = 'tab:red'
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)', color=color)
ax1.plot(x, y, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(which='major', axis='both', linestyle='--')

ax2 = ax1.twinx()  # fill in the blank - instantiate a second axes that shares the same x-axis

z = derivative(x)
color = 'tab:blue' #assigned variable color
ax2.set_xlabel('x')
ax2.set_ylabel('f(x)', color=color)
ax2.plot(x, z, color=color) #--- changed the y axis
ax2.tick_params(axis='y', labelcolor=color)
ax2.grid(which='major', axis='both', linestyle='--')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
