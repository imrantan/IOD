# Calculus - Limits

```python
import numpy as np
import matplotlib.pyplot as plt

# get matplot to show charts inline
%matplotlib inline
```

```python
# define a function, say f(x) = 1/x
def f(x):
    return 1/x
```

```python
# calculate value of f(x) for some values
print('f(4) = ',f(4))
```

```python
# calculate value of f(x) for 0
print('f(0) = ', f(0))
```

Plot y versus x as lines and/or markers.

> matplotlib.pyplot.plot

[matplotlib.pyplot.plot](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html)

```python
# plot f(x) from a small number to a large number
x = np.arange(0.1,100,1)
y = f(x)

_=plt.plot(x,y)

plt.show()
```

```python
x = np.arange(0.1,100,1)
y = f(x)

fig, ax = plt.subplots(figsize =(10, 6))

ax.plot(x,y)

plt.show()
```

Plot f(x) from a small number to a large number with logarithmic scale.

> Hint: plt.yscale('log')

```python
# plot f(x) from a small number to a large number with logarithmic scale
x = np.arange(0.1,100,1)
y = f(x)

_=plt.plot(x,y)
_=plt.yscale('log')

plt.show
```

```python
x = np.arange(0.1,100,1)
y = f(x)

fig, ax = plt.subplots(figsize =(8, 10))

_=plt.plot(x,y)
_=plt.yscale('log')

plt.show
```

```python
fig, ax = plt.subplots(figsize = (15,10))
#
ax.plot(x,y)
#
ax2 = ax.twinx()
ax2.plot(x,y)

_=plt.yscale('log')
plt.show()
```
