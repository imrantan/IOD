# 7. [BONUS] Plot X, Y and Best Fit Line

**Write a function to calculate best fit.**

![alt text](https://i.imgur.com/B43vOV0.jpg)[1]

[1]: https://www.mathsisfun.com/data/least-squares-regression.html "Least Squares Regression"

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for m and b
def best_fit(X, Y):
    '''
    Return m, b
    '''
    x1 = np.array(X)
    y1 = np.array(Y)      
    m = (len(x1)*np.sum(x1*y1) - np.sum(x1)*np.sum(y1)) / (len(x1)*np.sum(x1**2) - (np.sum(x1))**2) #gradient
    b = (np.sum(y1) - m*np.sum(x1)) / len(x1) #y-intercept
    
    return m, b

# solution
m, b = best_fit(X, Y)

print(m,b)

#best fit line:
#y = 0.80 + 0.92x

Y2 = m*np.array(X) + b

# plot points and fit line

plt.scatter(X,Y)
plt.plot(X,Y2,color='r')
plt.show()
````

# 11. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

````python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    # Calculate m, h
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a) #calculates the standard error of mean
    h = se * scipy.stats.t.ppf((1+confidence) / 2., n-1)
    return m, m-h, m+h

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
print(mean_confidence_interval(data))
````
