```python
#using plotly.express to plot an interactive boxplot. 
#Shows the distribution of the data of the male and female patient's ages.

import plotly.express as px

fig = px.box(heart_failure, x='sex', y='age', points="all") #the argument points="all" shows all the points.

fig.update_layout(title_text="Gender wise Age Spread - Male = 1 Female =0")

fig.show()

#Shows the 2 boxplots side by side for comparison as well as the distribution of the samples.
```

```python
#Styled box plot
#instead of sex now the x-axis is death event and the different colors represent smokers and non-smokers.

fig = px.box(heart_failure, x="DEATH_EVENT", y="age", color="smoking", notched=True)
fig.show()
```

```python
#using plotly.graph_objs to plot a donut chart to show how many patients survived after the follow up.

import plotly.graph_objs as go

#Step 1: Filter the data for male and female

male = heart_failure[heart_failure["sex"]==1] 
female = heart_failure[heart_failure["sex"]==0]

#Step 2: Filter the data for males & females who died and who are alive

male_survi = male[heart_failure["DEATH_EVENT"]==0]
male_not = male[heart_failure["DEATH_EVENT"]==1]
female_survi = female[heart_failure["DEATH_EVENT"]==0]
female_not = female[heart_failure["DEATH_EVENT"]==1]

#Step 3: Define the labels and values

labels = ['Male - Survived','Male - Not Survived', "Female -  Survived", "Female - Not Survived"]
values = [len(male[heart_failure["DEATH_EVENT"]==0]),len(male[heart_failure["DEATH_EVENT"]==1]),
         len(female[heart_failure["DEATH_EVENT"]==0]),len(female[heart_failure["DEATH_EVENT"]==1])]

#Step 4: Plot the donut

fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.4)])

fig.update_layout(title_text="Analysis on Survival - Gender")

fig.show()
```
